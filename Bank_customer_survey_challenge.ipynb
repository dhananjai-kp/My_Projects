```python
X = df_new.loc[:, ~df_new.columns.isin(['y'])]
Y = df_new.loc[:,'y']
```

```python
from sklearn.model_selection import train_test_split
# Split Training and Testing Data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,
                                test_size = 0.25, random_state=7)
```

```python
# NB Adaboost RF
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost
from sklearn import svm
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
```

```python
nb = GaussianNB()

rfc=RandomForestClassifier(n_estimators=100,random_state=7)

ab = AdaBoostClassifier(n_estimators=100, random_state=7)

xgb = xgboost.XGBClassifier()

svc = svm.SVC()

dtc = tree.DecisionTreeClassifier()

knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
def apply_model(clf,X_train,Y_train,X_test,Y_test):
    from sklearn.metrics import confusion_matrix,accuracy_score
    from sklearn.metrics import recall_score,precision_score,f1_score
    clf.fit(X_train,Y_train)
    Y_pred = clf.predict(X_test)
    cf_mat=confusion_matrix(Y_test,Y_pred)
    acc = accuracy_score(Y_test,Y_pred)
    prec = precision_score(Y_test,Y_pred)
    recall = recall_score(Y_test,Y_pred)
    f1 = f1_score(Y_test,Y_pred)
    return cf_mat, acc, prec , recall, f1
```

```python
for clf in [nb,rfc,ab,xgb,svc,dtc,knn]:
    cf_mat, acc, prec , recall, f1 = apply_model(clf,X_train,Y_train,X_test,Y_test)
    print(clf ,'\n', cf_mat, '  ', acc,'  ', prec ,'  ', recall,'  ', f1)
    print('-'*120)
```

Though Random Forest has more accuracy we can not consider it to be the best model because our data is highly imbalanced. Because the true positives are too many the accuracy will always be very high so we can not decide the model based on accuracy score.

So we need to choose decide based on recall and precision score. In our problem statement the False negative is more critical than false positive so we will decide the model based on recall score.

So Gaussian Naive Bayes with almost recall score 0.58 will be a better model
